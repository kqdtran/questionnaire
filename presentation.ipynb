**Import basic Python libraries** for use in the program: [os.path](http://docs.python.org/2/library/os.path.html) and [ConfigParser](http://docs.python.org/2/library/configparser.html).   
Also, import `re` for regular expression, and `numpy` + `pyplot` for plotting figures.   
Import `collections` since we are using `OrderedDict` for an ordered dictionary
Finally, import `gspread` for reading data from the Google Spreadsheet

```python
import os.path
import ConfigParser
import re
import numpy as np
import matplotlib.pyplot as plt
import collections
import gspread
```

**Read the username and password** from the `[google]` section of the `stat157.cfg` config file from your virtual machine home directory.

```python
configfile = os.path.join('.', 'stat157.cfg')
config = ConfigParser.SafeConfigParser()
config.read(configfile)
username = config.get('google', 'username')
password = config.get('google', 'password')
```

**Read the docid of the Google Spreadsheet** from the config file.

```python
docid = config.get('questionnaire', 'docid')
client = gspread.login(username, password)
spreadsheet = client.open_by_key(docid)
worksheet = spreadsheet.get_worksheet(0)
print docid
```

**Choose two columns from the spreadsheet to analyze**. We chose **learning style** and **previous experience**.

```python
LEARNING_STYLE = 'What is your learning style?'
EXPERIENCE = 'Have you had any personal experience in any of the following topics?'

fieldnames = [EXPERIENCE, LEARNING_STYLE]
print fieldnames
```

**Read in ALL rows of data from the Google Spreadsheet, but filter out columns that are not listed in `fieldnames`.**

```python
filtered_data = []
for row in worksheet.get_all_records():
    filtered_data.append({k:v for k,v in row.iteritems() 
                          if k in fieldnames})
print "Number of rows: {}".format(len(filtered_data))
```

* Iterate over the data, extract out the **experience** and put the items into a dictionary
* In the same iteration, use regex to strip the **scores** and the **learning style** of every students

```python
experienceDict = {}
scores = []  # Visual, Aural, Read/Write, Kinesthetic in that order
learningStyles = []

for row in filtered_data:
    experience = row[EXPERIENCE].split(',')  # split string by ','
    
    # Search for skill
    for skill in experience:
        skill = skill.strip().lower()  # strip whitespace
        if skill != 'none' and skill != 'no':
            if skill in experienceDict:
                experienceDict[skill] += 1
            else:
                experienceDict[skill] = 1
                
    # Search for learning style
    learningStyle = row[LEARNING_STYLE]
    currentScore = [int(s) for s in re.findall('[0-9]+', learningStyle)]
    if currentScore:
        scores.append(currentScore)
    style = [s.strip().lower() for s in 
             re.findall("(?<=You have a)(.*)(?=learning preference)", learningStyle)]
    if style:
        learningStyles.append(style[0])
    
experienceDict = collections.OrderedDict(sorted(experienceDict.items(), key=lambda (k, v): v))
```

**Plot the data generated above** using Matplotlib

```python
# Attach text labels at the end of the graph
def autolabel(rects):
    for ii,rect in enumerate(rects):
        height = rect.get_height()
        plt.text(height+rect.get_width(), rect.get_y()+0.2,
                '%s'% (experienceDict.values()[ii]),
                ha='center', va='bottom', size=14)

fig, ay = plt.subplots()
ind = np.arange(17)
width = 0.75
rects1 = ay.barh(ind, tuple(experienceDict.values()), width, alpha=0.5, align='center')
autolabel(rects1)

# Change size of whole graph
fig = plt.gcf()
fig.set_size_inches(13,8)
plt.savefig('experience.png', dpi=100)

plt.yticks(range(len(experienceDict)), experienceDict.keys(), size=18)
plt.xticks(size=18)
plt.xlabel('\nNumber of Students', size=20)
plt.title('Have you had any personal experience in any of the following topics??\n', size=25)

# Adjust graph limits
ymin, ymax = plt.ylim()   # return the current ylim
plt.ylim(ymax = 17, ymin=-1)  # set the ylim to ymin, ymax
plt.xlim(xmax = 35)

plt.tick_params(axis='y', right=False, left=False)
```

**Generate data for Learning Scores graph**

```python
N = len(scores)
visual = [s[0] for s in scores]
aural = [s[1] for s in scores]
read = [s[2] for s in scores]
kine = [s[3] for s in scores]
```

**Plot the Learning Scores data**

```python
ind = np.arange(N)    # the x locations for the groups
width = 0.8       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, visual, width, color='purple')
p2 = plt.bar(ind, aural, width, color='blue', bottom=visual)
p3 = plt.bar(ind, read, width, color='green', bottom=[aural[k]+visual[k] for k in range(N)])
p4 = plt.bar(ind, kine, width, color='violet', bottom=[aural[k]+visual[k]+read[k] for k in range(N)])

# Change size of whole graph
fig = plt.gcf()
fig.set_size_inches(15, 10)
plt.savefig('learningStyle.png', dpi=100)

# Labels and legends
plt.xlabel('Students', size=18)
plt.ylabel('Learning Style Scores', size=18)
plt.title('Learning Style Scores For Each Student', size=25)
plt.xticks(ind+width/2., ['S' + str(k) for k in range(N)])
plt.yticks(np.arange(0,50,2))
plt.legend((p1[0], p2[0], p3[0], p4[0]), ('Aural', 'Visual', 'Read/Write', 'Kinesthetic'))

plt.show()
```
